```python
# PLot
sns.set_style('whitegrid')
fig,ax = plt.subplots(2,2,figsize = (10,7))
sns.lineplot(data = flights, x = 'year', y ='passengers',ax=ax[0,0])
sns.lineplot(data=flights,x= 'month',y='passengers',ax=ax[0,1])
sns.lineplot(data=flights,x= 'year',y='passengers',hue='month',ax=ax[1,0])
sns.lineplot(data=flights,x= 'month',y='passengers',hue = 'year',ax=ax[1,1])

# Ticks setting
ax[0,0].set_xticks(flights['year'])
ax[0,0].tick_params(which = 'both',labelrotation=45,length = 4,color = 'r') 
ax[1,0].legend(prop={'size':7})
plt.tight_layout()
plt.show()
```
